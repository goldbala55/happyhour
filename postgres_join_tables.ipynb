```python
# Assign postgres users and connection strings
# dbname
dbname='happyhour'

# users
ro_user='happyhour_ro1'

# connection strings
ro_string = f"postgresql://{ro_user}:{ro_pass}@postgress-1.caahnjhin4st.us-east-1.rds.amazonaws.com/{dbname}"
```

### Using sqlalchemy

```python
# and now create the engine - use rw connection
engine = create_engine(ro_string)

# run a join 
join_sql = 'select t."PassengerId", pd."Name", t."title", pd."Sex", t."Sex", pd."SibSp", pd."Parch", t."Pclass", pd."Cabin" \
                        from passenger_details as pd \
                        inner join train as t on (t."PassengerId" = pd."PassengerId")'
        
join = engine.execute(join_sql)

for row in join:
    print(row)
```
