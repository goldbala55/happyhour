```python
# Import Libraries 
import pandas as pd 
import numpy as np 
import numpy
import matplotlib.pyplot as plt
# Install PyCaret
from pycaret.classification import *
```

```python
# Retrieve Data (csv)
train= pd.read_csv('Resources_2/train.csv',index_col= None)
test= pd.read_csv('Resources_2/test.csv',index_col= None)
```

```python
# Set up Environment
setup(data=train, target='Survived', ignore_features=['PassengerId', 'Name']
      , session_id=1, silent=True)
```

```python
# Compare Models
best_model = compare_models()
```

```python
print(best_model)
```

```python
#plot_model(best_model,plot='feature')
evaluate_model(best_model)
```

```python
# Finalize Model
finalized_model = finalize_model(best_model)
```

```python
# Predict
prediction = predict_model(finalized_model, data=test)
```

```python
model = pd.DataFrame({
        'PassengerId': test.PassengerId,
        'Survived': prediction.Label
    })
model.to_csv('model.csv', index=False)
```
